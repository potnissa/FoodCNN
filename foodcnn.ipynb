```python
from torch.utils.data import Dataset, DataLoader
import json
from PIL import Image
from torchvision.transforms import transforms

class Food101Dataset(Dataset):
    def __init__(self, json_path, root_dir):
        self.image_paths = []
        self.labels = []
        self.root_dir = root_dir

        full_json_path = os.path.join(root_dir, json_path)
        with open(full_json_path, 'r') as file:
            data = json.load(file)

        for food_name in data.keys():
            for image_filename in data[food_name]:
                self.image_paths.append(image_filename)
                self.labels.append(food_name)

        

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_path = os.path.join(self.root_dir, 'images', self.image_paths[idx] + '.jpg')
        image = Image.open(image_path).convert('RGB')
        print(image.size)
        transform = transforms.Compose([transforms.ToTensor()])


        tensor = transform(image)
       
        return tensor, self.labels[idx]
        
```

```python
root_dir = '/kaggle/input/food41'
data = Food101Dataset('meta/meta/train.json', root_dir)
train_dataloader = DataLoader(data, batch_size = 10, shuffle=True)
for batch in train_dataloader:
    print(batch)
```
